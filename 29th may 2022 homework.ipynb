```python
#q3 : Try to print all the number divisible by 3 in between a range of 40 - 400
a=40
b=400
while a<=b:
    if a%3==0:
        print(a,end=(" "))
    a+=1
```

```python
#q5 : Try to generate all the even number between 1- 1000
c=0
d=1000
while c<=d:
    if c%2==0:
        print(c,end=" ")
    c+=1
```
